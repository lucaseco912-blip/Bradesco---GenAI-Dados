GRAVAÇÃO DE ÁUDIO (Python + JavaScript)

```python
from IPython.display import Javascript
from google.colab import output
from base64 import b64decode
import io

def record_audio(filename="recorded.wav", duration=5):
    display(Javascript("""
    async function recordAudio(duration) {
      const stream = await navigator.mediaDevices.getUserMedia({ audio: true });
      const recorder = new MediaRecorder(stream);
      const chunks = [];

      recorder.ondataavailable = e => chunks.push(e.data);
      recorder.start();

      await new Promise(resolve => setTimeout(resolve, duration * 1000));
      recorder.stop();

      await new Promise(resolve => recorder.onstop = resolve);

      const blob = new Blob(chunks);
      const arrayBuffer = await blob.arrayBuffer();
      const base64 = btoa(
        new Uint8Array(arrayBuffer)
          .reduce((data, byte) => data + String.fromCharCode(byte), '')
      );

      google.colab.kernel.invokeFunction('notebook.save_audio', [base64], {});
    }

    recordAudio(%d);
    """ % duration))

    def save_audio(base64_audio):
        audio_bytes = b64decode(base64_audio)
        with open(filename, "wb") as f:
            f.write(audio_bytes)

    output.register_callback('notebook.save_audio', save_audio)

    print("🎙 Gravando por", duration, "segundos...")

record_audio("recorded.wav", 5)
```

Transcrever

```python
import whisper

model = whisper.load_model("small")

result = model.transcribe("recorded.wav", fp16=False, language=language)
transcription = result["text"]

print("🗣 Você disse:")
print(transcription)
```

INTEGRAÇÃO COM CHATGPT

```python
!pip install -q openai
```

Configurar API

```python
from openai import OpenAI

client = OpenAI(api_key="SUA_API_KEY_AQUI")
```

Enviar para GPT

```python
response = client.chat.completions.create(
    model="gpt-4o-mini",
    messages=[
        {"role": "system", "content": "Responda de forma clara e objetiva."},
        {"role": "user", "content": transcription}
    ]
)

chatgpt_response = response.choices[0].message.content

print("🤖 ChatGPT respondeu:")
print(chatgpt_response)
```

TRANSFORMAR RESPOSTA EM VOZ (gTTS)

```python
!pip install -q gTTS
```

Gerar áudio

```python
from gtts import gTTS
from IPython.display import Audio

tts = gTTS(text=chatgpt_response, lang=language, slow=False)

response_audio = "response_audio.wav"
tts.save(response_audio)

Audio(response_audio, autoplay=True)
```
